```python
#Assignment-2
#BMI Classification based on gender, height and weight using the machine learning
#Task 1 Data Exploration and clean up
# Importing the necessary library pandas to read and handle actions on csv files
import pandas as pd

# load the bmi dataset for BMI Classification
bmi_data = pd.read_csv('bmi.csv')
print(bmi_data.head())

# Summary of the dataset for basic statistics using describe
print("summary of dataset\n")
print(bmi_data.describe())

# Handling missing values by filling with the mean
bmi_data['Weight'] = bmi_data['Weight'].fillna(bmi_data['Weight'].mean())
bmi_data['Height'] = bmi_data['Height'].fillna(bmi_data['Height'].mean())

# Confirm missing values have been handled
print(bmi_data.isnull().sum())
```

```python
#Task2 Exploratory data analysis 
# Import libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Plot BMI category distribution usin thhe countplot
sns.countplot(x='Index', data=bmi_data)
plt.title('Distribution of BMI Categories')
plt.show()

# Visualizing the height and weight using hue as BMI category
palette = {0: "red", 1: "orange", 2: "yellow", 3: "green", 4: "blue", 5: "purple"}
sns.scatterplot(x='Height', y='Weight', hue='Index', data=bmi_data,palette=palette)
plt.title('Height and Weight by BMI Category')
plt.show()

# Gender specific distributions for BMI Index
sns.histplot(bmi_data[bmi_data['Gender'] == 'Male']['Index'], kde=True, label='Male', color='blue')
sns.histplot(bmi_data[bmi_data['Gender'] == 'Female']['Index'], kde=True, label='Female', color='green')
plt.title('BMI Category by Gender')
plt.legend()
plt.show()

# Plot Height distribution by gender male,female.
sns.histplot(bmi_data[bmi_data['Gender'] == 'Male']['Height'], kde=True, label='Male', color='blue')
sns.histplot(bmi_data[bmi_data['Gender'] == 'Female']['Height'], kde=True, label='Female', color='green')
plt.title('Height by Gender')
plt.legend()
plt.show()

# Plot Weight distribution by gender male, female.
sns.histplot(bmi_data[bmi_data['Gender'] == 'Male']['Weight'], kde=True, label='Male', color='blue')
sns.histplot(bmi_data[bmi_data['Gender'] == 'Female']['Weight'], kde=True, label='Female', color='green')
plt.title('Weight by Gender')
plt.legend()
plt.show()

# Box Plot of Height distribution by BMI Category
# Helps identify height variations across different BMI categories
sns.boxplot(x='Index', y='Height', data=bmi_data)
plt.title('Height Distribution across BMI Categories')
plt.xlabel('BMI Category')
plt.ylabel('Height')
plt.show()

# Box Plot of Weight distribution by BMI Category
# Helps identify weight variations across different BMI categories
sns.boxplot(x='Index', y='Weight', data=bmi_data)
plt.title('Weight Distribution across BMI Categories')
plt.xlabel('BMI Category')
plt.ylabel('Weight')
plt.show()

# Shows correlations between numerical features (Height, Weight, BMI Index)
plt.figure(figsize=(8, 6))
correlation_matrix = bmi_data[['Height', 'Weight', 'Index']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#Task3 DATA Preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# assume genders like Male as 1, Female as 0
bmi_data['Gender'] = bmi_data['Gender'].map({'Male': 1, 'Female': 0})

# Define features and target variable
features = bmi_data[['Gender', 'Height', 'Weight']]
target = bmi_data['Index']

# Addressing class imbalance using SMOTE
smote = SMOTE()
X_balanced, y_balanced = smote.fit_resample(features, target)

# Scaling features for better model performance
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X_balanced)

# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y_balanced, test_size=0.3, random_state=42)
print("done the data preprocessing")
```

```python
#Task 4 Model training and prediction
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Logistic Regression training and prediction 
logistic = LogisticRegression()

logistic.fit(X_train, y_train)
y_logistic_predict = logistic.predict(X_test)

# Support Vector Machine training and prediction
svm = SVC()
svm.fit(X_train, y_train)
y_svm_predict = svm.predict(X_test)

# K-Nearest Neighbors training and prediction
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_knn_predict = knn.predict(X_test)

print("Logistic Regression, SVM, and KNN models trained successfully.")
```

```python
#Task 5 Model evaluation and comparision
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import recall_score, f1_score

# Logistic Regression Performance or evaluating logistic model using the metrics like accuracy etc.
logistic_accuracy = accuracy_score(y_test, y_logistic_predict)
logistic_precision = precision_score(y_test, y_logistic_predict, average='weighted', zero_division=1)
logistic_recall = recall_score(y_test, y_logistic_predict, average='weighted', zero_division=1)
logistic_f1 = f1_score(y_test, y_logistic_predict, average='weighted')

print("Logistic Regression Performance using sklearn metrics")
print(f"Accuracy: {logistic_accuracy}, Precision: {logistic_precision}, Recall: {logistic_recall}, F1 Score: {logistic_f1}")
print("\n")
# SVM Performance
svm_accuracy = accuracy_score(y_test, y_svm_predict)
svm_precision = precision_score(y_test, y_svm_predict, average='weighted', zero_division=1)
svm_recall = recall_score(y_test, y_svm_predict, average='weighted', zero_division=1)
svm_f1 = f1_score(y_test, y_svm_predict, average='weighted')

print("SVM Performance using the sklearn metrics")
print(f"Accuracy: {svm_accuracy}, Precision: {svm_precision}, Recall: {svm_recall}, F1 Score: {svm_f1}")
print("\n")

# KNN Performance
knn_accuracy = accuracy_score(y_test, y_knn_predict)
knn_precision = precision_score(y_test, y_knn_predict, average='weighted', zero_division=1)
knn_recall = recall_score(y_test, y_knn_predict, average='weighted', zero_division=1)
knn_f1 = f1_score(y_test, y_knn_predict, average='weighted')

print("KNN Performance using the sklearn metrics")
print(f"Accuracy: {knn_accuracy}, Precision: {knn_precision}, Recall: {knn_recall}, F1 Score: {knn_f1}")
```

```python
#Task 6 Gender based prediction analysis
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_female_data = bmi_data[bmi_data['Gender'] == 0][['Height', 'Weight']]
y_female_data = bmi_data[bmi_data['Gender'] == 0]['Index']

X_male_data = bmi_data[bmi_data['Gender'] == 1][['Height', 'Weight']]
y_male_data = bmi_data[bmi_data['Gender'] == 1]['Index']
#now standardize the male data using the standard scler 
scaler=StandardScaler()
x_male_scale=scaler.fit_transform(X_male_data)
x_male_train,x_male_test,y_male_train,y_male_test=train_test_split(x_male_scale,y_male_data,test_size=0.3, random_state=42)

#now standardize the female data using the StandardScaler
X_female_scale=scaler.fit_transform(X_female_data)
x_female_train,x_female_test,y_female_train,y_female_test=train_test_split(X_female_scale,y_female_data,test_size=0.3,random_state=42)

# Train the Logistic Regression for males
log_male_model = LogisticRegression()
log_male_model.fit(x_male_train, y_male_train)
y_male_pred_log = log_male_model.predict(x_male_test)

# Evaluate Logistic Regression for males
accuracy_male_log = accuracy_score(y_male_test, y_male_pred_log)
precision_male_log = precision_score(y_male_test, y_male_pred_log, average='weighted', zero_division=1)
recall_male_log = recall_score(y_male_test, y_male_pred_log, average='weighted', zero_division=1)
f1_male_log = f1_score(y_male_test, y_male_pred_log, average='weighted')

print("Logistic Regression Model Performance for Male:")
print(f"Accuracy: {accuracy_male_log}, Precision: {precision_male_log}, Recall: {recall_male_log}, F1 Score: {f1_male_log}")
print("\n")

# Train Logistic Regression for females
log_female_model = LogisticRegression()
log_female_model.fit(x_female_train, y_female_train)
y_female_pred_log = log_female_model.predict(x_female_test)

# Evaluate Logistic Regression for females
accuracy_female_log = accuracy_score(y_female_test, y_female_pred_log)
precision_female_log = precision_score(y_female_test, y_female_pred_log, average='weighted', zero_division=1)
recall_female_log = recall_score(y_female_test, y_female_pred_log, average='weighted', zero_division=1)
f1_female_log = f1_score(y_female_test, y_female_pred_log, average='weighted')

print("Logistic Regression Model Performance for Female:")
print(f"Accuracy: {accuracy_female_log}, Precision: {precision_female_log}, Recall: {recall_female_log}, F1 Score: {f1_female_log}")
print("\n")
```

```python
# SVM model for Males
svm_model_male = SVC()
svm_model_male.fit(x_male_train, y_male_train)
y_male_pred_svm = svm_model_male.predict(x_male_test)

# Evaluate SVM model for males
accuracy = accuracy_score(y_male_test, y_male_pred_svm)
precision = precision_score(y_male_test, y_male_pred_svm, average='weighted', zero_division=1)
recall = recall_score(y_male_test, y_male_pred_svm, average='weighted', zero_division=1)
f1_male = f1_score(y_male_test, y_male_pred_svm, average='weighted')

print("SVM Model Performance for Male:")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1_male}")
print("\n")


# SVM model for Females
svm_model_female = SVC()
svm_model_female.fit(x_female_train, y_female_train)
y_female_pred_svm = svm_model_female.predict(x_female_test)

# Evaluate SVM model for females
accuracy = accuracy_score(y_female_test, y_female_pred_svm)
precision = precision_score(y_female_test, y_female_pred_svm, average='weighted', zero_division=1)
recall = recall_score(y_female_test, y_female_pred_svm, average='weighted', zero_division=1)
f1_female = f1_score(y_female_test, y_female_pred_svm, average='weighted')

print("SVM Model Performance for Female:")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1_female}")

print("\n")
```

```python
# KNN model for Males
knn_model_male = KNeighborsClassifier(n_neighbors=5)
knn_model_male.fit(x_male_train, y_male_train)
y_male_pred_knn = knn_model_male.predict(x_male_test)
# Evaluate KNN model for males
accuracy = accuracy_score(y_male_test, y_male_pred_knn)
precision = precision_score(y_male_test, y_male_pred_knn, average='weighted', zero_division=1)
recall = recall_score(y_male_test, y_male_pred_knn, average='weighted', zero_division=1)
f1 = f1_score(y_male_test, y_male_pred_knn, average='weighted')

print("KNN Model Performance for Male:")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

print("\n")


# KNN model for Females
knn_model_female = KNeighborsClassifier(n_neighbors=5)
knn_model_female.fit(x_female_train, y_female_train)
y_female_pred_knn = knn_model_female.predict(x_female_test)

print("KNN Model Performance for Female:")

# Evaluate KNN model for females
accuracy = accuracy_score(y_female_test, y_female_pred_knn)
precision = precision_score(y_female_test, y_female_pred_knn, average='weighted', zero_division=1)
recall = recall_score(y_female_test, y_female_pred_knn, average='weighted', zero_division=1)
f1 = f1_score(y_female_test, y_female_pred_knn, average='weighted')

print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1 Score: {f1}")

print("\n")
```
