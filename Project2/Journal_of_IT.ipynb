```python
#Journal of Information Technology
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
url='https://journals.sagepub.com/toc/jina/39/4'
with open('Journal_of_IT.csv',mode='a',newline='',encoding='utf-8') as file:
    writer=csv.writer(file)
    writer.writerow(['Title','Issue_Title','Issue_URL'])
def findIssues(url):
    driver=webdriver.Chrome()
    driver.get(url)
    try:
        title=driver.find_element(By.CSS_SELECTOR,'div.spd__title').text.strip()
    except Exception as e:
        title='NA'
   
    issues_info=driver.find_elements(By.CSS_SELECTOR,'div.issue-item__title')
    for issue in issues_info:
        issue_ele=issue.find_element(By.CSS_SELECTOR,'a')
        issue_url=issue_ele.get_attribute('href')
        issue_title=issue_ele.text.strip()
        with open('Journal_of_IT.csv',mode='a',newline='',encoding='utf-8') as file:
            writer=csv.writer(file)
            writer.writerow([title,issue_title,issue_url])
    try:
        prev=driver.find_element(By.CSS_SELECTOR,'a.content-navigation__btn--pre')
        prev_url=prev.get_attribute('href')
        driver.quit()
        findIssues(prev_url)
    except Exception as e:
        print('No more previous element')
    
findIssues(url)

    
```

```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
journal_data=pd.read_csv('Journal_of_IT.csv')
with open('Journal_of_IT_articles.csv',mode='a',newline='',encoding='utf-8') as file:
    writer=csv.writer(file)
    writer.writerow(['Title','Volume','Year','URL','Author','Author_email','Author_data'])
for index,row in journal_data.iterrows():
    driver=webdriver.Chrome()
    volume=row['Title'].split(',')[0]
    year=row['Title'].split(',')[1]
    issue_title=row['Issue_Title']
    url=row['Issue_URL']
    driver.get(url)
    try:
        authors_info=driver.find_element(By.CSS_SELECTOR,'a.to-authors-affiliations')
        authors_info.click()
        time.sleep(2)
        print("authors data")
        hide=driver.find_element(By.CSS_SELECTOR,'button.collateral-contributors-control')
        if hide.get_attribute('aria-expanded')=='false':
            hide.click()
            time.sleep(2)
        authors=driver.find_elements(By.CSS_SELECTOR, 'div[id^="con"]')
        for auth in authors:
            auth_title=auth.find_element(By.CSS_SELECTOR,'div.heading h4').text.strip()
            print(auth_title)
            try:
                auth_data=auth.find_element(By.CSS_SELECTOR,'div.affiliations').text.strip()
            except Exception as e:
                auth_data='NA'
            try:
                auth_email=auth.find_element(By.CSS_SELECTOR,'div.core-email').text.strip()
            except Exception as e:
                auth_email='NA'
            with open('Journal_of_IT_articles.csv',mode='a',newline='',encoding='utf-8') as file:
                writer=csv.writer(file)
                writer.writerow([issue_title,volume,year,url,auth_title,auth_email,auth_data])
        
    except Exception as e:
        print('exception',e)
        with open('Journal_of_IT_articles.csv',mode='a',newline='',encoding='utf-8') as file:
            writer=csv.writer(file)
            writer.writerow([issue_title,volume,year,url,'NA','NA','NA'])
    driver.quit()
            
        
    
```
