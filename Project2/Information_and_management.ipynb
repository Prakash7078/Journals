```python
#for both information and management and Information and organization same journals for aericles so using same code with different start url.
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import csv
with open('information_and_organization.csv',mode='a',newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Volume_Issue','Vol_Year','ArticleUrl','Article_Title'])  # Write the header row
def findVolumeIssues(url): 
    global count
    vol_issue,vol_year='N/A','N/A'
    if not url:
        return 0
    driver=webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(20)
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'h2.js-vol-issue'))
        )
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'h3.js-issue-status'))
        )
        vol_issue=driver.find_element(By.CSS_SELECTOR,'h2.js-vol-issue').text.strip()
        vol_year=driver.find_element(By.CSS_SELECTOR,'h3.js-issue-status').text.strip()
    except Exception as e:
        print(e)
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'li.js-article-list-item'))
        )
        articles=driver.find_elements(By.CSS_SELECTOR,'li.js-article-list-item')
        for article in articles:
            url=article.find_element(By.CSS_SELECTOR,'a.article-content-title')
            article_url=url.get_attribute('href')
            title=url.text.strip()
            with open('information_and_organization.csv',mode='a',newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow([vol_issue,vol_year,article_url,title])  # Write the header row
                file.flush()
    except Exception as e:
        print(e)
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.navigation-pre'))
        )
        prev=driver.find_element(By.CSS_SELECTOR,'div.navigation-pre')
        prev_url_element=prev.find_element(By.CSS_SELECTOR,'a.anchor')
        prev_url=prev_url_element.get_attribute('href')
        driver.quit()
        return findVolumeIssues(prev_url)
    except Exception as e:
        print(e)
findVolumeIssues('https://www.sciencedirect.com/journal/information-and-organization/vol/34/issue/4')   

```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import csv
import time

#initalize the chrome webdriver

driver=webdriver.Chrome()
journals_data=pd.read_csv('information_and_organization.csv')
journal_urls=journals_data['ArticleUrl']
print(journal_urls)
# driver.get('https://www.sciencedirect.com/science/article/pii/0167923685901915')
# # Wait for the page to fully load
# driver.implicitly_wait(20)
# author_group=driver.find_elements(By.CSS_SELECTOR,'div.author-group')
# if author_group:
#     print('author group')
# else:
#     print('no author group')
def getAuthorsData(authors,driver):
    authdata=[]
    for auth in authors:
        name,desc='',''
        try:
            auth.click()
            # Wait for the side panel to open
            time.sleep(2)  
            auth_data=driver.find_element(By.CSS_SELECTOR,'div.side-panel-content')
            auth_data_desc=auth_data.find_element(By.CSS_SELECTOR,'div.affiliation')
            auth_name=auth_data.find_element(By.CSS_SELECTOR,'div.author')
            name=auth_name.text.strip()
            desc=auth_data_desc.text.strip()
        except Exception as e:
            print("error at author data",e)
        try:
            auth_email=auth_data.find_element(By.CSS_SELECTOR,'div.e-address a.anchor')
            email=auth_email.text.strip()
        except Exception as e:
            email='N/A'
        authdata
        authdata.append([name,email,desc])
    return authdata
        




for url in journal_urls:
    driver=webdriver.Chrome()
    final_data=[]
    driver.get(url)
    # Wait for the page to fully load
    driver.implicitly_wait(10)
    # Wait for the page to load using WebDriverWait
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'h1.Head'))
        )
        title=driver.find_element(By.CSS_SELECTOR,'h1.Head').text.strip()
        article_details=driver.find_element(By.CSS_SELECTOR,'div.Publication')
        article_journal=article_details.find_element(By.CSS_SELECTOR,'h2.publication-title').text.strip()
        article_vol=article_details.find_element(By.CSS_SELECTOR,'div.text-xs a.anchor').text.strip()
                
        article_month_year=article_details.find_element(By.CSS_SELECTOR,'div.text-xs').text.split(',')[2]
        keywords=driver.find_elements(By.CSS_SELECTOR,'div.keyword')
        abstract=driver.find_element(By.CSS_SELECTOR,'div.u-margin-s-bottom').text.strip()
        keyword_list=[]
        for key in keywords:
            keyword=key.text.strip()
            keyword_list.append(keyword)
    except Exception as e:
        print(f"Error is {e} in this {url}")
        
    final_data=[url,article_journal,title,article_vol,article_month_year]
    final_data.append(abstract)
    final_data.append(keyword_list)
    try:
        author_group=driver.find_element(By.CSS_SELECTOR,'div.author-group')
        authors=author_group.find_elements(By.CSS_SELECTOR,'button.button-link')
        auth_data=getAuthorsData(authors,driver)
        for i in auth_data:
            with open('information_and_organization_journal_articles.csv', mode='a', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow(final_data + i)
                file.flush()
    except Exception as e:
            with open('information_and_organization_journal_articles.csv', mode='a', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow(final_data + ["N/A", "N/A", "N/A"])
                file.flush()
            print(f"Error processing author data on {url}: {e}")
    driver.quit()


driver.quit()
```
